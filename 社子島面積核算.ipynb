```python
import pandas as pd
import numpy as np
```

```python
shezi = pd.read_csv('C:/Users/10909282/Desktop/社子島地號編定_1.csv',encoding='big5')
```

```python
shezi.head()
```

```python
shezi['combine'] = shezi['SecID_Land']+'_'+shezi['分區子號']
```

```python
a = shezi.groupby('combine').sum()
a = a.reset_index()
a
```

```python
b = a['combine'].str.split("_")
b
```

```python
sec_land = []
land_type = []
for i in range(len(b)):
    sec_land.append(b[i][0])
    land_type.append(b[i][1])
```

```python
a['sec_land']=sec_land
a['land_type']=land_type
```

```python
a
```

```python
x = a[['Shape_Area']]
y = a[['sec_land']]
z = a[['land_type']]
```

```python
area_sum = pd.concat([y,z,x],axis=1)
area_sum
```

```python
area_sum = area_sum.to_csv("area_sum.csv",encoding="utf-8",index=False)
```
