```python
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
owner = pd.read_excel('C:/Users/jerry/Desktop/土城所有權人整理.xlsx',sheet_name='土城所有權人整理1')
owner.head()
```

```python
owner = owner.sort_values(by=['登記日期'],ascending=False,ignore_index=True)
```

```python
a = owner.groupby('地段號').count()
a = a.reset_index()
land_id = a['地段號']
land_id
```

## 建新DF儲存所有權人資料

```python
own_df = pd.DataFrame(columns=['地段號','所有權人組合','姓名組合','持分比組合'])
own_df['地段號']=land_id
own_df
```

```python
def owner_combination(sec_id):
    for i in range(0,len(owner)):
        if owner.loc[i]['地段號']==sec_id:
            name1 = owner[owner['地段號']==sec_id]['姓名_持分'].values
            name2 = owner[owner['地段號']==sec_id]['姓名'].values
            name3 = owner[owner['地段號']==sec_id]['百分比(未約分)'].values
            string = ','.join('%s' %id for id in name1)
            names = ','.join('%s' %id for id in name2)
            ratio = ','.join('%s' %id for id in name3)
            return string,names,ratio
```

```python
for i in range(len(land_id)):
    own_df['所有權人組合'][i],own_df['姓名組合'][i],own_df['持分比組合'][i] = owner_combination(own_df['地段號'][i])
    print(own_df['所有權人組合'][i])
```

```python
for i in range(len(own_df)):
    own_df['持分比組合'][i] = own_df['持分比組合'][i].replace(" ","")
```
