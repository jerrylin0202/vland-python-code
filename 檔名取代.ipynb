```python
import pandas as pd
import os
```

```python
path = '//192.168.0.252//徵收查估部//03.查估部一般案件//110年度//VL-C-11010002土城司法園區區段徵收開發案委託專業服務案//03_查估資料//05_清冊//各所有權人清冊PDF//公示 - 複製//0.農暫存'

file_list = next(os.walk(path))[1]
file_list
```

```python
def change_file_name(file_path):
    ori_name = next(os.walk(path+'//'+file_path))[2]
    data_list = next(os.walk(path+'//'+file_path))[2]
    for i in range(len(data_list)):
        data_list[i] = data_list[i].replace('神明','O')
        data_list[i] = data_list[i].replace('營業設備遷移獎勵金','N')
        data_list[i] = data_list[i].replace('生產設備遷移獎勵金','M')
        data_list[i] = data_list[i].replace('農附','L')
        data_list[i] = data_list[i].replace('體育','K')
        data_list[i] = data_list[i].replace('水井','J')
        data_list[i] = data_list[i].replace('房屋補助費','I')
        data_list[i] = data_list[i].replace('人口電話瓦斯','H')
        data_list[i] = data_list[i].replace('營業損失','G')
        data_list[i] = data_list[i].replace('設備搬遷','F')
        data_list[i] = data_list[i].replace('農固','E')
        data_list[i] = data_list[i].replace('畜禽農機PVC管','D')
        data_list[i] = data_list[i].replace('林','C')
        data_list[i] = data_list[i].replace('農','B')
        old_file_name = path+'//'+file_path+'//'+ori_name[i]
        new_file_name = path+'//'+file_path+'//'+data_list[i]
        try:
            print('舊檔案名稱:'+ori_name[i]+'轉換為:'+data_list[i])
            os.rename(old_file_name,new_file_name)
        except Exception as e:
            print(e)
            pass
```

```python
for i in range(len(file_list)):
    try:
        print(file_list[i])
        change_file_name(file_list[i])
    except:
        pass
```
